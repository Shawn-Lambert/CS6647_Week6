```python
# The starting state is Rest. Calculate the probability of possible activity on the 15th day. 

import numpy as np
import random as rm

# All possible states
states = ["Rest","Sweets","Exercise"]

# All state transitions
transitionName = [["RR","RE","RS"],["ER","EE","ES"],["SR","SE","SS"]]

# Probabilities of state transitions
transitionMatrix = [[0.2,0.6,0.2],[0.1,0.6,0.3],[0.2,0.7,0.1]]

if sum(transitionMatrix[0])+sum(transitionMatrix[1])+sum(transitionMatrix[1]) != 3:
    print("Sum of all state transition weighted edges do not add up to 3. Check probabilities of state transitions matrix.")
else: print("All state transition probabilities add up. Continue.")

# Markov model function to forecast state transition based on given days
def activity_forecast(days):
    # Starting state
    activityToday = "Rest"
    print("Start state: " + activityToday)
    # List to store state transitions
    activityList = [activityToday]
    i = 0
    # To calculate the probability of the activityList
    prob = 1
    while i != days:
        if activityToday == "Rest":
            change = np.random.choice(transitionName[0],replace=True,p=transitionMatrix[0])
            if change == "RR":
                prob = prob * 0.2
                activityList.append("Rest")
                pass
            elif change == "RE":
                prob = prob * 0.6
                activityToday = "Exercise"
                activityList.append("Exercise")
            else:
                prob = prob * 0.2
                activityToday = "Sweets"
                activityList.append("Sweets")
        elif activityToday == "Exercise":
            change = np.random.choice(transitionName[1],replace=True,p=transitionMatrix[1])
            if change == "EE":
                prob = prob * 0.5
                activityList.append("Exercise")
                pass
            elif change == "ER":
                prob = prob * 0.2
                activityToday = "Rest"
                activityList.append("Rest")
            else:
                prob = prob * 0.3
                activityToday = "Sweets"
                activityList.append("Sweets")
        elif activityToday == "Sweets":
            change = np.random.choice(transitionName[2],replace=True,p=transitionMatrix[2])
            if change == "SS":
                prob = prob * 0.1
                activityList.append("Sweets")
                pass
            elif change == "SR":
                prob = prob * 0.2
                activityToday = "Rest"
                activityList.append("Rest")
            else:
                prob = prob * 0.7
                activityToday = "Exercise"
                activityList.append("Exercise")
        i += 1  
    print("Possible states: " + str(activityList))
    print("End state after "+ str(days) + " days: " + activityToday)
    print("Probability of the possible sequence of states: " + str(prob))

# Markov model function that forecasts state based on ‘days’ parameter
activity_forecast(15)
```

```python
#What about the state and probability after 10000 days?

# Markov model function that forecasts state based on ‘days’ parameter
activity_forecast(10000)
```

```python
# What do you observe from the above two?

# The following probabilities are based on entering the exercise state:
# From ‘Rest’ to ‘Exercise’ = 0.6
# From ‘Sweets’ to ‘Exercise’ = 0.7
# From ‘Exercise’ to ‘Exercise’ = 0.6
# These probabilities are much higher than the probability of entering any other state.
# Therefore, at the end of any given number of days there is a high probability that the state will be ‘Exercise’.
```
